```python
class Account:
    def __init__(self, owner: str, balance: float = 0.0):
        self.owner = owner
        if balance < 0:
            print(f"ê²½ê³ : ì´ˆê¸° ì”ì•¡ì€ ìŒìˆ˜ê°€ ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ì”ì•¡ì„ 0ìœ¼ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.")
            self._balance = 0.0
        else:
            self._balance = balance

    def deposit(self, amount: float):
        if amount <= 0:
            print("ğŸš« ì…ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")
            return

        self._balance += amount
        print(f"ğŸ’° {amount:.0f}ì› ì…ê¸ˆ ì™„ë£Œ. í˜„ì¬ ì”ì•¡: {self._balance:.0f}ì›")

    def withdraw(self, amount: float):
        if amount <= 0:
            print("ğŸš« ì¶œê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")
            return

        if self._balance - amount < 0:
            print(f"âŒ ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self._balance:.0f}ì›, ìš”ì²­ ì¶œê¸ˆì•¡: {amount:.0f}ì›")
            return

        self._balance -= amount
        print(f"ğŸ’¸ {amount:.0f}ì› ì¶œê¸ˆ ì™„ë£Œ. í˜„ì¬ ì”ì•¡: {self._balance:.0f}ì›")

    def get_balance(self) -> float:
        return self._balance

    def __str__(self) -> str:
        return f"ê³„ì¢Œ ì†Œìœ ì£¼: {self.owner}, í˜„ì¬ ì”ì•¡: {self._balance:.0f}ì›"
    
my_account = Account("ë°•ìƒì¤€", 100000)

print("--- ê³„ì¢Œ ìƒì„± ë° ì´ˆê¸° ìƒíƒœ ---")
print(my_account)

print("\n--- ì…ê¸ˆ ë° ì¶œê¸ˆ ì‹œë®¬ë ˆì´ì…˜ ---")
my_account.deposit(50000)
my_account.withdraw(30000)
my_account.withdraw(150000)
my_account.deposit(-10000)
my_account.withdraw(0)

print("\n--- ìµœì¢… ì”ì•¡ ì¶œë ¥ ---")
print(f"'{my_account.owner}'ë‹˜ì˜ ìµœì¢… ì”ì•¡ì€ {my_account.get_balance():.0f}ì›ì…ë‹ˆë‹¤.")
print(my_account)
```
