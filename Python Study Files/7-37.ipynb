[ math 모듈 ]  
 - 수학과 관련된 기능을 가지고 있음
math 모듈 내 함수  
 - sin(x) : 사인 값을 구함
 - cos(x) : 코사인 값을 구함
 - tan(x) : 탄젠트 값을 구함

 - floor(x) : 내림  
  
이 외에도 광장히 많음....

```python
import math

print(math.sin(5))
print(math.cos(5))
print(math.ceil(3.6))
print(math.floor(4.1))
```

[ from 구문 ]  
모듈안에 수 많은 함수 중 우리가 원하는 함수만을 가져와 사용하면 문법적으로 굉장히 편해집니다.  
from 모듈 이름 import 가져오고 싶은 변수 또는 함수  


---

from math import sin  
(기존) math.sin(5)  
(from 구문 사용 후) sin(5)  

```python
from math import sin, cos, ceil, floor

print(sin(5))
print(cos(5))
print(ceil(3.6))
print(floor(4.1))
```

```python
# math 모듈의 모든 함수를 가져오는 방법으로 import에 '*' 사용합니다.
from math import *
```

[ as 구문 ]  
모듈을 가져오다 보면 이름의 충돌이 발생하는 경우가 있습니다. 또는 너무 길어 짧게 줄여서 사용하고 싶은 경우도 있을겁니다.  
이럴때 모듈의 식별자를 만들어 사용할 수 있는 문법이 as 구문입니다.

```python
import math as m

print(m.sin(5))
print(m.cos(5))
print(m.ceil(3.6))
print(m.floor(4.1))
```

[ os 모듈 ]  
 - 운영체제와 관련된 기능을 가진 모듈
 - os.system 명령어의 경우 사용 시 주의를 요함

```python
import os

print('현재 운영체제 : ', os.name)
print('현재 폴더 : ', os.getcwd())
print('현재 운영체제 : ', os.listdir())

# 폴더를 만들고
os.mkdir('oz')
# 폴더를 삭제하고
os.rmdir('oz')

with open('oz.txt', 'w') as file:
  file.write('수강생 여러분 안녕하세요')
os.rename('oz.txt', 'oz2.txt')

# 파일 제거
os.remove('oz2.txt')

os.system('ls') # 맥
os.system('dir') # 윈도우
```

[ datetime 모듈 ]  
 - 날짜와 시간 관련된 모듈

```python
import datetime

print('오늘 날짜와 시간 출력')
now = datetime.datetime.now()
print(now.year, '년')
print(now.month, '월')
print(now.day, '일')
print(now.hour, '시')
print(now.minute, '분')
print(now.second, '초')
```

```python
import datetime

print('다양한 시간 포맷')
output_1 = now.strftime('%Y.%m.%d %H:%M:%S')
output_2 = f'{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초'
output_3 = now.strftime('%Y{} %m{} %d{} %H{} %M{} %S{}').format(*'년월일시분초') # * 기호 때문에 나누어져서 들어감.

print(output_1)
print(output_2)
print(output_3)
```
