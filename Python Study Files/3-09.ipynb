```python
 # 시퀀스 자료형에 다 지원되는 메소드
 번호1, 번호2, 번호3, 번호4 = input().split(",")

 번호1 = int(번호1)
 번호2 = int(번호2)
 번호3 = int(번호3)
 번호4 = int(번호4)

 print(type(번호1), 번호2, 번호3, 번호4)
```

```python
# map()
번호1, 번호2, 번호3, 번호4 = map(int, input().split(","))

print(type(번호1), 번호2, 번호3, 번호4)
```

```python
# 레거시 : 누군가 남겨두고간 코드
# 리팩토링 : 결과의 변경 없이 코드의 구조를 재조정하는 것입니다.
# input()

fish__p, fish__s, fish__j = 0, 0, 0

fish__p, fish__s, fish__j = map(int, input("붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)를 순서대로 개수를 입력하고 꼭 쉼표를 입력하고\n그리고 안사는 붕어빵에는 0을 넣어주세요.").split(","))
fish__p__sell, fish__s__sell, fish__j__sell = 2000, 2500, 3000
order__p = fish__p * fish__p__sell
order__s = fish__s * fish__s__sell
order__j = fish__j * fish__j__sell

total__ = fish__p + fish__s + fish__j
total__pay = order__p + order__s + order__j

print(f'주문이 완료되었습니다. 팥 붕어빵 {fish__p}개, 슈크림 붕어빵 {fish__s}개, 잡채 붕어빵 {fish__j}개 총 {total__}개 결제 금액은 {total__pay}원 입니다.')
```
