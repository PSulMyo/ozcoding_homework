[ 데코레이터 ]  
 - @으로 시작하는 구문을 데코레이터라고 함
 - 데코레이터는 만드는 방법에 따라 함수 데코레이터와 클래스 데코레이터로 나눌 수 있음  
  
[ 함수 데코레이터 ]  
함수에 사용되는 데코레이터는 함수 앞뒤에 붙일 내용이나 반복할 내용이 있을때 사용합니다.

```python
def oz():
  print('파이썬 진짜 재미있엉 하하')

oz()
```

```python
def test(function):
  def wrapper():
    print('허언증이 재발했습니다.')
    function()
    print('격리 되었습니다.')
  return wrapper

@test
def oz():
  print('파이썬 진짜 재미있엉 하하')

oz()
```

웹 프레임워크에서 굉장히 많이 사용되는 기능입니다.  
데코레이터의 기능과 사용방법에 대한 내용을 더 찾아보고 공부해보시길 바랍니다.
