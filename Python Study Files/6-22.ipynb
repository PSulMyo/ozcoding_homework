```python
# 매개변수를 적게 입력했을때
def oz(message, n):
  for i in range(n):
    print(message)

oz('oz코딩스쿨', 5, 3)
```

```python
# def 함수이름(매개변수, *매개변수)
def oz(n, *values):
  # 코드
  for i in range(n):
    for value in values:
      print(value)
    print()

oz(3, '이렇게', '저렇게', '가변 매개변수는 요렇게', '신기하죠')
```

```python
# def 함수이름(매개변수, *매개변수)
def oz(*values, n):
  # 코드
  for i in range(n):
    for value in values:
      print(value)
    print()

oz('이렇게', '저렇게', '가변 매개변수는 요렇게', '신기하죠', 3)
```
