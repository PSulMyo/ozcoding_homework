##시퀀스 자료형
* 문자열 :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
    * 슬라이싱 ( [:] [: :])
    * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등...)
* list :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
    * 슬라이싱 ( [:] [: :])
    * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등...)
    * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* Tuple :
    * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
    * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 많다.
    * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
    * 그리고 엄청난 비밀이 숨겨져있어요
    * 튜플은 소괄호 입니다.
* dict(딕셔너리)
    * list, tuple, str, range(범위안에 연속된 데이터를 만드는 시퀀스)  
    공통적인 특징이 있어요 데이터의 연관성이 없습니다.
    * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재  
    (예시 : 게임 캐릭터 힘:30, 지능:20, 체력:40, 민첩:90)
    * dict 생성방법은 중괄호

* set
    * set은 수학의 집합을 의미합니다.
    * 딕셔너리와 같이 순서가 없습니다.
    * 값의 중복을 허락하지 않습니다. 아주 중요합니다.
    * set 생성 방법도 중괄호


```python
채소 = {"당근", "배추", "대파", "양파"}
type(채소)
print(채소)

# 순서가 없고, 출력하면 요소가 다르게 나오는것도 특징입니다.

# 세트를 만드는 다양한 방법
세트 = set()
세트
세트2 = {}
세트2
```

```python
# 중복을 제일 싫어하는 set

사과_리스트 = list("apple")
print(사과_리스트)
사과_튜플 = tuple("apple")
print(사과_튜플)
사과_셋 = set("apple")
print(사과_셋)
```

```python
# set = 집합, 합집합, 교집합, 차집합 기타 등등 !
# 대표적으로 합집합, 교집합, 차집합

채소1 = {"당근", "양파", "오이", "배추"}
채소2 = {"양파", "오이", "대파", "가지"}

# 합집합(union)
채소 | 채소2
set.union(채소1, 채소2)

# 교집합(intersection)
채소1 & 채소2
set.intersection(채소1, 채소2)

# 차집합(difference)
채소1 - 채소2
set.difference(채소1, 채소2)
```
