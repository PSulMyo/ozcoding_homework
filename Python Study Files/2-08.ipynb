```python
# 중요
# == : 값만 비교
# is : 개체의 타입을 비교

3 is 3
3 == 3

3.0 is 3
3.0 == 3

print(id(3.0))
print(id(3))
```

```python
# 본격적인 논리연산자
# True : 1 / False : 0
True and True # 1 * 1 = 1
True and False # 1 * 0 = 0
False and True # 0 * 1 = 0
False and False # 0 * 0 = 0
```

```python
# 본격적인 논리연산자
# True : 1 / False : 0
True or True # 1 + 1 = 1
True or False # 1 + 0 = 1
False or True # 0 + 1 = 1
False or False # 0 + 0 = 0
```

```python
#not

not True # False

not False # True
```

```python
# 점점 복잡해지는 조건들!!!

10 == 12 or 6 != 2 # True
2 > 5 and 7 == 7 # False
not 14 <= 15 # False
not 17 is 7.0 # True
```
