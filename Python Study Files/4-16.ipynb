```python
# 3. while문 조건식 증감, 감소

# 1. 기본적인 증가 유형
i = 1
while i < 10:
  print('이정도로 성장할줄이야', i)
  i += 1


# 2. 기본적인 감소 유형
i = 10
while i > 0:
  print('이정도로 성장할줄이야', i)
  i -= 1
```

```python
# 5. while문이 for문 보다 강점인 부분?
# 무수히 많이 실행되어야 할때

import random

# random.randint(1,6)

i = 0
while i != 4:
  i = random.randint(1,6)
  print(i)
```

```python
# 6. while문에도 else가 있어요

양파 = 10
while 양파 > 10:
  print("양파", 양파, "개 남았습니다.")
  if 양파 == 0:
    break
  양파 -= 1
else:
  print("양파 다 팔렸습니다.")
```

```python
# 7. 반복문에 흐름을 제어하는 continue와 break

i = 0
while True:
  print(i)
  i += 1
  if i == 10:
    break

for i in range(100):
  print(i)
  if i == 10:
    break
```

```python
i = 0
while i < 10:
  i += 1
  if i % 2 == 0:
    continue
  print(i)

for i in range(10):
  if i % 2 == 0:
    continue
  print(i)
```

```python
# 8. 붕어빵 인터네셔널은 내가 살린다 나도 이제 개발자
# for문을 이용해 리팩토링을 해주세요.
# 잡담 게시판에 여러분의 코드를 자랑해주세요. 깔끔한 코드 상 1명 / 노력 상 1명 스벅 쿠폰을 보내드리겠습니다.

bungeoppang_types = ["팥", "슈크림", "잡채"]
bungeoppang_prices = [2000, 2500, 3000]

order_input_str = input(f"{'/'.join(bungeoppang_types)} 순서대로 드시고 싶은 붕어빵의 개수를 적어주세요. 구분은 ' , ' 로 하겠습니다.")
order_counts_str = order_input_str.split(',')
order_counts = list(map(int, order_counts_str))

total_order_quantity = 0
total_revenue = 0

item_total_costs = []

print("\n--- 붕어빵 주문 상세 내역 ---")

for i in range(len(bungeoppang_types)):
  bungeoppang_type = bungeoppang_types[i]
  price = bungeoppang_prices[i]
  order_qty = order_counts[i]

  item_cost = price * order_qty
  item_total_costs.append(item_cost)

  total_order_quantity += order_qty
  total_revenue += item_cost

  if order_qty > 0:
      print(f"- {bungeoppang_type} 붕어빵: {order_qty}개 (개당 {price}원) -> 총 {item_cost}원")

print("--------------------------")
print(f"총 주문 붕어빵 개수: {total_order_quantity}개")
print(f"총 지불하실 금액: {total_revenue}원")
print("--------------------------")

```

```python
from ast import Num
# 9. 예제로 while문 익혀보기

# 콜라츠하는 사람은 주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들 수 있다고 추측했다.
# 작업
# 1-1. 입력된 수가 짝수라면 2로 나눈다.
# 1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
# 1-3. 위 결과로 나온 수에 같은 작업을 1이 될 때까지 반복한다.
# 예를 들어, 입력된 수가 6이라면, 6->3->10->5->16->8->4->2->1

i = int(input("숫자를 입력하세요: "))

while i > 1:
  print(int(i))
  if i % 2 == 0:
    i = i / 2
  else:
    i = i * 3 + 1

```
