```python
def my_fancy_generator(): # ì œë„ˆë ˆì´í„° í•¨ìˆ˜ ì •ì˜
    print("--- ë§ˆë²• ì§€íŒ¡ì´ë¥¼ ë“¤ì—ˆìŠµë‹ˆë‹¤! ---")
    yield "ì§ !"
    print("--- ë‘ ë²ˆì§¸ ë§ˆë²• ì¤€ë¹„ ---")
    yield "ë¿…!"
    print("--- ë§ˆë²• ë ---")

# 1ë‹¨ê³„: ì œë„ˆë ˆì´í„° í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ì œë„ˆë ˆì´í„° ê°ì²´ ìƒì„±!
# ì´ë•Œ ì•„ì§ 'ì§ !'ì´ë‚˜ 'ë¿…!'ì€ ì¶œë ¥ë˜ì§€ ì•Šì•„ìš”!
gen_obj = my_fancy_generator()
print("ğŸ‰ ì œë„ˆë ˆì´í„° ê°ì²´ê°€ ìƒì„±ë˜ì—ˆì–´ìš”!")


# 2ë‹¨ê³„: next() í•¨ìˆ˜ë¡œ ê°’ì„ í•˜ë‚˜ì”© ì–»ì–´ì™€ ë§ˆë²• ë°œë™!

value1 = next(gen_obj)
print(f"ì²« ë²ˆì§¸ ë§ˆë²•: {value1}")
# ì¶œë ¥:
# --- ë§ˆë²• ì§€íŒ¡ì´ë¥¼ ë“¤ì—ˆìŠµë‹ˆë‹¤! ---
# ì²« ë²ˆì§¸ ë§ˆë²•: ì§ !

value2 = next(gen_obj)
print(f"ë‘ ë²ˆì§¸ ë§ˆë²•: {value2}")
# ì¶œë ¥:
# --- ë‘ ë²ˆì§¸ ë§ˆë²• ì¤€ë¹„ ---
# ë‘ ë²ˆì§¸ ë§ˆë²•: ë¿…!

# ë” ì´ìƒ yieldê°€ ì—†ìœ¼ë©´ StopIteration ë°œìƒ!
# try-exceptë¡œ ì˜ˆì™¸ë¥¼ ì²˜ë¦¬í•´ì£¼ê±°ë‚˜, for ë£¨í”„ë¥¼ ì‚¬ìš©í•˜ëŠ” ê²ƒì´ ì¼ë°˜ì ì´ì—ìš”.
try:
    value3 = next(gen_obj)
    print(f"ì„¸ ë²ˆì§¸ ë§ˆë²•: {value3}")
except StopIteration:
    print("â›”ï¸ ë” ì´ìƒ ë§ˆë²•ì€ ë‚˜ì˜¤ì§€ ì•ŠìŠµë‹ˆë‹¤. ì œë„ˆë ˆì´í„°ê°€ ëë‚¬ì–´ìš”!")
# ì¶œë ¥:
# --- ë§ˆë²• ë ---
# â›”ï¸ ë” ì´ìƒ ë§ˆë²•ì€ ë‚˜ì˜¤ì§€ ì•ŠìŠµë‹ˆë‹¤. ì œë„ˆë ˆì´í„°ê°€ ëë‚¬ì–´ìš”!
```
