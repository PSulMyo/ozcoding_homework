```python
def test():
  print('제너레이터 실습 코드입니다.')
  yield 'test'

print('첫번째 출력')
test()

print('두번째 출력')
test()

print(test())
```

```python
def test():
  print('첫번째 출력')
  yield 1
  print('두번째 출력')
  yield 2
  print('세번째 출력')

output = test()
print('네번째 출력')

x = next(output)
print(x)
print('다섯번째 출력')
y = next(output)
print(y)
print('여섯번째 출력')
z = next(output)
print(z)
```

```python
# - 이터레이터와 제너레이터에 대해 설명해주세요.
# ㄴ 이터레이터는 데이터를 순서대로 하나씩 꺼낼 수 있는 객체 /
#  for 루프를 사용해서 리스트나 문자열을 반복할 수 있는 것은 내부적으로 이터레이터를 제공하기 때문.
# ㄴ 제너레이터는 이터레이터를 생성해 주는 특별한 함수 /
#  일반 함수와는 달리 return 대신 yield 키워드를 사용함, yield 만나면 값을 돌려주고
#  함수 실행을 일시 중단했다가, 다음에 next() 가 호출되면 중단되었던 부분부터 이어서 실행됨.
# - 제너레이터 함수를 만드는 방법에 대해 설명해주세요.
# ㄴ 일반 함수처럼 정의(def()), yield 키워드 사용(함수 안에 하나 이상의 yield 키워드를 포함하면 됨.)
# - 제너레이터 함수를 실행하는 방법에 대해 설명해주세요.
# 1. next() 함수를 호출하면 제너레이터 객체는 제너레이터 함수 안의 코드를 yield 키워드를 만날 때까지 실행.
# 2. yield 를 만나면, 해당 값을 반환하고 함수 실행을 일시 중단함. 이전에 어디까지 실행했는지,
#  어떤 변수 값을 가졌는지 모두 기억한 채로 멈췄던 부분부터 다시 이어서 실행해요.
# 📌 예외 : 더 이상 yield 가 남아있지 않거나, 함수 끝까지 실행되면 StopIteration 이라는 예외(에러)를 발생시켜서
#  더 이상 줄 값이 없음을 알려줌.
```

```python
def my_fancy_generator(): # 제너레이터 함수 정의
    print("--- 마법 지팡이를 들었습니다! ---")
    yield "짠!"
    print("--- 두 번째 마법 준비 ---")
    yield "뿅!"
    print("--- 마법 끝 ---")

# 1단계: 제너레이터 함수를 호출하여 제너레이터 객체 생성!
# 이때 아직 '짠!'이나 '뿅!'은 출력되지 않아요!
gen_obj = my_fancy_generator()
print("🎉 제너레이터 객체가 생성되었어요!")


# 2단계: next() 함수로 값을 하나씩 얻어와 마법 발동!

value1 = next(gen_obj)
print(f"첫 번째 마법: {value1}")
# 출력:
# --- 마법 지팡이를 들었습니다! ---
# 첫 번째 마법: 짠!

value2 = next(gen_obj)
print(f"두 번째 마법: {value2}")
# 출력:
# --- 두 번째 마법 준비 ---
# 두 번째 마법: 뿅!

# 더 이상 yield가 없으면 StopIteration 발생!
# try-except로 예외를 처리해주거나, for 루프를 사용하는 것이 일반적이에요.
try:
    value3 = next(gen_obj)
    print(f"세 번째 마법: {value3}")
except StopIteration:
    print("⛔️ 더 이상 마법은 나오지 않습니다. 제너레이터가 끝났어요!")
# 출력:
# --- 마법 끝 ---
# ⛔️ 더 이상 마법은 나오지 않습니다. 제너레이터가 끝났어요!
```
