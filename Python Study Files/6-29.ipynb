```python
x, y = 1, 2

print('x : ', x)
print('y : ', y)
print()

x, y = y, x
print('x : ', x)
print('y : ', y)
```

```python
def oz_tuple():
  return (1, 2)

x, y = oz_tuple()

print('x : ', x)
print('y : ', y)
```

```python
# 튜플을 이용한 내부함수들
# enumerate()의 1, fruit는 (1, fruit)이나 괄호를 생략한 상태입니다.

for i, fruit in enumerate([1, 2, 3, 4]):
  print(i, fruit)
```

```python
# divmode() 사용 시 튜플 형태로 몫과 나머지를 변환
a, b = 7, 3
x, y = divmod(a, b)
print('몫 : ', x)
print('나머지 : ', y)
```
