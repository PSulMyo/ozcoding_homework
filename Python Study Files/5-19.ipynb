```python
# 예외코드 작성

try:
  x = int(input('나눌 숫자를 입력하세요'))
  y = 10 / x
  print(y)
except:
  print("예외가 발생했습니다.")
```

```python
y = [10, 20, 30]
try:
  index, x = map(int, input('나눌 숫자를 입력하세요. : ').split(','))
  print(y[index] / x)
except:
  print('오류입니다.')
```

```python
y = [10, 20, 30]
try:
  index, x = map(int, input('나눌 숫자를 입력하세요. : ').split(','))
  print(y[index] / x)
except ZeroDivisionError:
  print('숫자를 0으로 나눌 수 없습니다.')
except:
  print('잘못된 인덱스입니다.')
```

```python
try:
  x = int(input('나눌 숫자를 입력하세요.'))
  y = 10 / x
  print(y)
except:
  print("예외가 발생합니다.")
else:
  print("성공하셨습니다.")
finally:
  print("코드 실행이 종료되었습니다.")
```
