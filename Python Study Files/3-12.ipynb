##시퀀스 자료형
* 문자열 :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
    * 슬라이싱 ( [:] [: :])
    * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등...)
* list :
    * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등...)
    * 슬라이싱 ( [:] [: :])
    * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등...)
    * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())
* Tuple :
    * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
    * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 많다.
    * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값, 값)
    * 그리고 엄청난 비밀이 숨겨져있어요
    * 튜플은 소괄호 입니다.
* dict(딕셔너리)
    * list, tuple, str, range(범위안에 연속된 데이터를 만드는 시퀀스)  
    공통적인 특징이 있어요 데이터의 연관성이 없습니다.
    * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재  
    (예시 : 게임 캐릭터 힘:30, 지능:20, 체력:40, 민첩:90)
    * dict 생성방법은 중괄호

```python
# list [], tuple (), dict {}

짱짱맨 = {"힘":30, "지능":20, "체력":40, "민첩":90}
짱짱맨

# {키 : 값}
# {key1 : value1, key2 : value2}

# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾았습니다.

# 짱짱맨[1]
짱짱맨['힘']
print(f'짱짱맨님은 힘이 {짱짱맨["힘"]}입니다')
```

```python
# 딕셔너리를 만드는 다양한 방법

# 빈 딕셔너리 만들기
딕셔너리 = {}
딕셔너리

딕셔너리2 = dict()
딕셔너리2

# 값을 넣어 만드는 방법
# dict()로 생성시 값을 넣는 기호 "", : 주의해주세요
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첩=90)
딕셔너리3

#zip() : 병렬처리를 도와줍니다.
딕셔너리4 = dict(zip(["힘" ,"지능", "체력", "민첩"], [30,20,40,90]))
딕셔너리4
```

```python
# 값의 중복은 허락되는데, 키의 중복은 허락하지 않습니다.

짱짱맨 = {"힘":30, "지능":20, "체력":40, "민첩":90, "힘":40, "지구력":90}
짱짱맨

"체력" in 짱짱맨
```

```python
# 딕셔너리에 다양한 메소드

print(짱짱맨.keys())
print(짱짱맨.values())

print(짱짱맨.items())
print(type(짱짱맨.items()))

리스트화된_짱짱맨 = list(짱짱맨.items())
리스트화된_짱짱맨[0][1]
```

* 해시, 해시앱, 해시테이블

```python
붕어빵 = ("팥 붕어빵", "슈크림 붕어빵", "잡채 붕어빵")
붕어빵_가격 = {
    "팥" : 2000,
    "슈크림" : 2500,
    "잡채" : 3000
}
팥, 슈크림, 잡채 = map(int, input("붕어빵(팥), 붕어빵(슈크림), 붕어빵(잡채)를 순서대로 개수를 입력하신 후 꼭 쉼표를 입력하고\n그리고 안사는 붕어빵에는 0을 넣어주세요.").split(","))
팥_붕어빵_총_가격 = 붕어빵_가격["팥"] * 팥
슈크림_붕어빵_총_가격 = 붕어빵_가격["슈크림"] * 슈크림
잡채_붕어빵_총_가격 = 붕어빵_가격["잡채"] * 잡채

총_가격 = 팥_붕어빵_총_가격 + 슈크림_붕어빵_총_가격 + 잡채_붕어빵_총_가격

print(f'주문이 완료되었습니다. {붕어빵[0]} {팥}개 총 {팥_붕어빵_총_가격}원,\n{붕어빵[1]} {슈크림}개 총 {슈크림_붕어빵_총_가격}원\n{붕어빵[2]} {잡채}개 총 {잡채_붕어빵_총_가격}원\n총합 {총_가격}원 입니다.')
```
