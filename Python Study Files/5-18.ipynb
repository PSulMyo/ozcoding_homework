```python
class 클래스이름:
  def __init__(self, 매개변수1, 매개변수2):
    self.속성1 = 매개변수1
    self.속성2 = 매개변수2
```

```python
class Person:
  def __init__(self, name, age, address):
    self.hello = '안녕하세요'
    self.name = name
    self.age = age
    self.address = address

  def greeting(self):
    print('{0} 저는 {1}입니다.'.format(self.hello, self.name))

maria = Person('아리아', 20, '서울시 서초구 반포동')
maria.greeting()    # 안녕하세요. 저는 아리아입니다.

print("이름:", maria.name)    # 아리아
print("나이:", maria.age)     # 20
print("주소:", maria.address) # 서울시 서초구 반포동
```

##클래스의 상속


```python
class Car(object):
  maxspeed = 300
  maxpeople = 5
  def move(self, x):
    print('스피드로 달리고 있습니다.')
  def stop(self):
    print('멈췄습니다.')

class HybridCar(Car):
  battery = 1000
  batteryKM = 300

class ElectricCar(HybridCar):
  battery = 1000
  batteryKM = 300
  def move(self, x):
    print(self.batteryKM, '만큼 달릴 수 있습니다.')
    print(x, '스파드로 달리고 있습니다.')

k5 = HybridCar()
k5.maxspeed

electricCark5 = ElectricCar()
electricCark5.maxspeed
electricCark5.battery
electricCark5.move(10)

# print(dir(k5))
```
