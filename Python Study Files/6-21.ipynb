```python
# 함수 선언부
def oz():
  for i in range(5):
    print('oz코딩스쿨')

# 함수 실행부
oz()
```

```python
print('어떤가요 이것도 함수였어요')

# print()만 입력해서 확인해보세요
print()
```

```python
# def 함수이름(매개변수, 매개변수):
def oz(message, n):
  # 코드
  for i in range(n):
    print(message)

oz('oz코딩스쿨', 5)
```
