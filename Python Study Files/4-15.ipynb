## 2. for문 문법과 주의사항(콜론, 들여쓰기)

```python
for 동물 in ("토끼","사자","푸바오","호랑이"):
  print(동물)
  print("보러 갈래요?")
```

```python
# 3. for문 만들어보기

for i in range(5):
  print("반복이 되네 신기해!!")
```

```python
# 4. for문의 흐름 확인해보면 좋겠죠?

for i in range(5):
  print("반복이 되네 신기해", i)
```

```python
# 7. 예제로 for문 익혀보기
# 문제 : 절대값 정수가 담긴 정수 변수와 그 정수의 부호를 차례 담은 부호 변수가 있다
# 두 변수를 이용해 실제 정수의 합을 구해주세요 (True : 양수, False : 음수)

정수 = [10, 5, 8]
부호 = [True, False, True]

total_sum = 0

for i in range(len(정수)):
  abs_val = 정수[i]
  sign = 부호[i]

  if sign:
    current_num = abs_val
  else:
    current_num = -abs_val

  total_sum += current_num

print(f'주어진 정수들의 합은 : {total_sum}')
```
